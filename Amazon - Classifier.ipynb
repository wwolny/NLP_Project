```python
import pandas as pd
import pickle
```

```python
dtm = pd.read_pickle('dtm_text_2.pkl')
df = dtm.transpose()
df.head()
```

```python
nb_model = pd.DataFrame(index = dtm.index)
for index, col in enumerate(dtm.columns):
    rows =[]
    for i, rank in enumerate(dtm.rating):
        rows.append((dtm.iloc[i,index]+1)/(sum(dtm.iloc[:,rank])+dtm.shape[1]))
    nb_model[col] = rows
```

```python
row = []
sum_rank = 0
for index, rank in enumerate(dtm.rating):
    row.append(sum(dtm.iloc[index,:]))
    sum_rank += sum(dtm.iloc[index,:])

for i in range(0, len(row)):
    row[i] = row[i]/sum_rank
    
nb_model["probability_score$"] = row
```
