## Explaratory Analysis

```python
import pandas as pd
import pickle
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
```

```python
dtm = pd.read_pickle('dtm_text_1.pkl')
df = dtm.transpose()
df_clean = pd.read_pickle('data_clean_1.pkl')
df_clean.head()
dtm.head()
```

```python
top_dict = {}
for c in df.columns:
    top = df[c].sort_values(ascending=False).head(50)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```

```python
# Print the top 15 words in each review
for rating, top_words in top_dict.items():
    print(rating)
    print(', '.join([str(word) for word, count in top_words[0:14]]))
    print('---')
```

```python
from collections import Counter
words = []
for rating in df.columns:
    for (word, count) in top_dict[rating]:
        words.append(word)

Counter(words).most_common()
```

```python
add_stop_words = [word for word, count in Counter(words).most_common() if count > 2]
len(add_stop_words)
```

```python
for index, col in enumerate(dtm.columns):
    if sum(dtm.iloc[:, index]) < 4:
        add_stop_words.append(col)
```

```python
# Let's update our document-term matrix with the new list of stop words
dtm = dtm.drop(add_stop_words, axis=1)
dtm.head()
```

```python
stop_words =nltk.corpus.stopwords.words('english')
stop_words.extend(add_stop_words)
data_clean_stop = pd.DataFrame()
data_clean_stop["rating"] = df_clean["rating"].unique()
rows = []
for index in range(0, len(df_clean["rating"])):
    rows.append(" ".join([word for word in df_clean.iloc[index,1].split() if word not in stop_words]))
    print(index)
data_clean_stop["text"] = rows
data_clean_stop.head()
```

```python
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
              max_font_size=150, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,6]

# rating_name = ['neutral', 'very good', 'good', 'very bad', 'bad']
rating_name = ['very good', 'good', 'very bad', 'bad']
for index, rating in enumerate(data_clean.columns):
    wc.generate(data_clean.text[index])
    plt.subplot(2,2, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis('off')
    plt.title(rating_name[index])

plt.show()
```
