# Sentiment Analysis

```python
import pandas as pd
from textblob import TextBlob
import numpy as np
import math
```

```python
data = pd.read_pickle('data_clean.pkl')
data
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

data['polarity'] = data['text'].apply(pol)
data['subjectivity'] = data['text'].apply(sub)
data
```

```python
long_data = pd.read_pickle('corpus_divided.pkl')
long_data.head()
```

```python
long_data['polarity'] = long_data['text'].apply(pol)
long_data['subjectivity'] = long_data['text'].apply(sub)
long_data.head()
```

```python
for rating in long_data.rating.unique():
#     print(long_data.where(long_data.rating == rating).polarity.sum(axis=0, skipna=True))
    polar = long_data.where(long_data.rating == rating).polarity.sum(axis=0, skipna=True)
    subject = long_data.where(long_data.rating == rating).subjectivity.sum(axis=0, skipna=True)
    size = long_data.where(long_data.rating == rating).count()
    print("polarity of " +str(rating) +": ")
    print(polar/size.rating)
    print("subjectivity of " +str(rating) +": ")
    print(subject/size.rating)
```

```python
title = pd.read_pickle("title.pkl")
title.head()
```

```python
title['polarity'] = title['title'].apply(pol)
title['subjectivity'] = title['title'].apply(sub)
title.head()
```

```python
for rat in title.rating.unique():
    polar = title[title["rating"]==rat].polarity.sum(axis=0, skipna=True)
    subject = title[title["rating"]==rat].subjectivity.sum(axis=0, skipna=True)
    size = title[title["rating"]==rat].count()
    print("polarity of " +str(rat) +": ")
    print(polar/size.rating)
    print("subjectivity of " +str(rat) +": ")
    print(subject/size.rating)
```
